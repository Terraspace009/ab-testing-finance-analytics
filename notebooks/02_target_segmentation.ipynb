```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("customers.csv")

# Check first few rows
df.head()
```

```python
# Summary statistics
df.describe()

# Check for missing values
df.isnull().sum()

# Class balance
df['signed_up'].value_counts(normalize=True)
```

```python
# Summary statistics
df.describe()

# Check for missing values
df.isnull().sum()

# Class balance
df['signed_up'].value_counts(normalize=True)
```

```python
# Age distribution
sns.histplot(df['age'], kde=True)
plt.title("Age Distribution")
plt.show()

# Income distribution
sns.histplot(df['income'], kde=True)
plt.title("Income Distribution")
plt.show()

# Spending Score
sns.histplot(df['spending_score'], kde=True)
plt.title("Spending Score Distribution")
plt.show()

# Risk Score
sns.countplot(x='risk_score', data=df)
plt.title("Risk Score Counts")
plt.show()
```

```python
# Signup by age
sns.boxplot(x='signed_up', y='age', data=df)
plt.title("Signup vs Age")
plt.show()

# Signup by income
sns.boxplot(x='signed_up', y='income', data=df)
plt.title("Signup vs Income")
plt.show()

# Signup by spending score
sns.boxplot(x='signed_up', y='spending_score', data=df)
plt.title("Signup vs Spending Score")
plt.show()

# Signup by risk score
sns.boxplot(x='signed_up', y='risk_score', data=df)
plt.title("Signup vs Risk Score")
plt.show()
```
